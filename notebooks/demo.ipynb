El dataset contiene una mezcla de variables numéricas y categóricas, además de valores nulos, lo que lo hace ideal para demostrar las capacidades del paquete.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from toolbox_ml.eda.core import (
    describe_df,
    tipifica_variables,
    get_features_num_regression,
    plot_features_num_regression,
    # get_features_cat_regression,    # disponible tras merge de Dev 3
    # plot_features_cat_regression,   # disponible tras merge de Dev 3
    # detect_outliers,                # disponible tras merge de Dev 3
)
```

## 1. Carga del dataset

```python
# Cargamos el dataset Titanic desde seaborn
df = sns.load_dataset('titanic')

print(f"Dimensiones: {df.shape}")
df.head()
```

## 2. `describe_df` — Resumen estadístico del DataFrame

Genera una vista rápida de cada columna: tipo de dato, porcentaje de nulos, valores únicos y cardinalidad.

Útil como primer paso de cualquier análisis exploratorio.

```python
resumen = describe_df(df)
resumen
```

```python
# Comprobación de entrada incorrecta → debe devolver None
describe_df("esto no es un dataframe")
```

## 3. `tipifica_variables` — Sugerencia de tipo estadístico

Clasifica cada columna en: **Binaria**, **Categórica**, **Numérica Discreta** o **Numérica Continua**,
en función de su cardinalidad y los umbrales que definimos.

```python
# umbral_categoria=10: columnas con menos de 10 únicos → categórica
# umbral_continua=30: si el % de cardinalidad supera 30% → continua
tipos = tipifica_variables(df, umbral_categoria=10, umbral_continua=30)
tipos
```

```python
# Comprobación de entrada incorrecta → debe devolver None
tipifica_variables(df, umbral_categoria=-1, umbral_continua=30)
```

## 4. `get_features_num_regression` — Selección de features numéricas

Devuelve las columnas numéricas cuya correlación de Pearson con el target supera el umbral indicado.

Opcionalmente filtra también por significancia estadística (p-valor).

```python
# Target: 'fare' (precio del billete)
# Umbral de correlación: 0.3
features_num = get_features_num_regression(df, target_col='fare', umbral_corr=0.3)
print("Columnas correlacionadas con 'fare':", features_num)
```

```python
# Con filtro adicional de p-valor
features_num_pval = get_features_num_regression(
    df, target_col='fare', umbral_corr=0.3, pvalue=0.05
)
print("Con p-valor < 0.05:", features_num_pval)
```

```python
# Comprobación de entrada incorrecta → debe devolver None
get_features_num_regression(df, target_col='columna_inexistente', umbral_corr=0.3)
```

## 5. `plot_features_num_regression` — Visualización de features numéricas

Pinta pairplots de las columnas numéricas correlacionadas con el target.
Si hay más de 5 columnas, las divide automáticamente en grupos.

```python
columnas_representadas = plot_features_num_regression(
    df, target_col='fare', umbral_corr=0.3, pvalue=0.05
)
print("Columnas representadas:", columnas_representadas)
```
